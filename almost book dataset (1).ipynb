```python
@interact
def recommended_book_publishers(publisher_name = list(df['publisher'].value_counts().index)):
    filtered_books = df[df['publisher'] == publisher_name][['title', 'average_rating']]
    sorted_books = filtered_books.sort_values(by='average_rating', ascending=False)
    return sorted_books.head(10)
```

```python
@interact
def recommended_book_authors(authors_name = list(df['authors'].value_counts().index)):
    filtered_books = df[df['authors'] == authors_name][['title', 'average_rating']]
    sorted_books = filtered_books.sort_values(by='average_rating', ascending=False)
    return sorted_books.head(10)
```

```python
@interact
def recommended_book_languages(languages = list(df['language_code'].value_counts().index)):
    filtered_books = df[df['language_code'] == languages][['title', 'average_rating']]
    sorted_books = filtered_books.sort_values(by='average_rating', ascending=False)
    return sorted_books.head(10)
```

```python
def num_to_obj(x):
    if x>0 and x<=1:
        return"between 0 and 1"
    if x>1 and x<=2:
        return "between 1 and 2"
    if x>2 and x<=3:
        return "between 2 and 3"
    if x>3 and x<=4:
        return "between 3 and 4"
    if x>4 and x<=5:
        return "between 4 and 5"
df['rating_obj'] = df['average_rating'].apply(num_to_obj)    
```

```python
df['rating_obj'].value_counts()
```

```python
rating_df = pd.get_dummies(df['rating_obj'])
rating_df.head()
```

```python
language_df = pd.get_dummies(df['language_code'])
language_df.head()
```

```python
features = pd.concat([rating_df,language_df,df['average_rating'],df['ratings_count'],df['title']],axis=1)
features.set_index('title', inplace=True)
features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
features_scaled
```

```python
from sklearn import neighbors
```

```python
model = neighbors.NearestNeighbors(n_neighbors=6, algorithm = 'ball_tree',metric = 'euclidean')
model.fit(features_scaled)
dist, idlist = model.kneighbors(features_scaled)
```

```python
@interact
def BookRecommender(book_name = list(df['title'].value_counts().index)):
    book_list_name =[]
    book_id = df[df['title'] == book_name].index
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list_name.append(df.loc[newid].title)
    return book_list_name
```
